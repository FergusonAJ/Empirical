```python
import pandas as pd
loaded_files = []
headers = ["Out0{}".format(i) for i in range(10)]
for i in range(10):
    moo = pd.read_csv("./runs/out0{}".format(i), header=None, index_col=0)
    loaded_files.append(moo)

squish = pd.concat(loaded_files, axis=1)
squish.columns = headers
squish
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(style="white", palette="muted", color_codes=True)

sns.distplot(squish["Out00"], bins=10)
sns.distplot(squish["Out01"], bins=10)
sns.distplot(squish["Out02"], bins=10)
sns.distplot(squish["Out03"], bins=10)
```
